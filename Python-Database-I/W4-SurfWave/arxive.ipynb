```python
import socket
import urllib
from BeautifulSoup import *
```

```python
url1 = "https://arxiv.org/find/all/1/all:+AND+Rajamani+narayanan/0/1/0/all/0/1"
url2 = "https://arxiv.org/find/all/1/all:+AND+Rajamani+narayanan/0/1/0/all/0/1?skip=25&query_id=c7436199a24b1a29"
url3 = "https://arxiv.org/find/all/1/all:+AND+Rajamani+narayanan/0/1/0/all/0/1?skip=50&query_id=c7436199a24b1a29"
url4 = "https://arxiv.org/find/all/1/all:+AND+Rajamani+narayanan/0/1/0/all/0/1?skip=75&query_id=c7436199a24b1a29"

html = urllib.urlopen(url1).read()\
       +urllib.urlopen(url2).read()\
       +urllib.urlopen(url3).read()\
        +urllib.urlopen(url4).read()
    
soup = BeautifulSoup(html)
tags = soup('span')
ass = soup('a')
```

```python
n=1
i =0
stop =0
jpt = [0,'pdf','ps','other']
Authors = []
AUT = []
for a in ass:
    if a.contents[0]== 'other':
        
        #------------------------------------- 
        aut1 = str(ass[n].contents[0])
        Authors.append(aut1)
        #------------------------------------- 
        aut2 = str(ass[n+1].contents[0])
        if aut2[0:5] == 'arXiv':
            arxiv = aut2
            Authors.append(arxiv)
        else:
            Authors.append(aut2)
        #-------------------------------------     
        aut3 = str(ass[n+2].contents[0])
        if aut3 in jpt:
            stop = 1
        else:
            if aut3[0:5] == 'arXiv':
                arxiv = aut3
                Authors.append(arxiv)
            else:
                Authors.append(aut3)
        #-------------------------------------  
        if stop != 1:
            aut4 = str(ass[n+3].contents[0])
        
            if aut4 in jpt:
                stop = 1
            else:
                if aut4[0:5] == 'arXiv':
                    arxiv = aut4
                    Authors.append(arxiv)
                else:
                    Authors.append(aut4)
        #-------------------------------------  
        if stop != 1:
            aut5 = str(ass[n+4].contents[0])
       
            if aut5 in jpt:
                stop = 1
            else:
                if aut5[0:5] == 'arXiv':
                    arxiv = aut5
                    Authors.append(arxiv)
                else:
                    Authors.append(aut5)
        
        #------------------------------------- 
        AUT.append(Authors)
        Authors = []
        stop = 0
    n = n+1
```
